# Melodix — Exploration Notebook
Explore songs and test recommendation models.

```python
import sys; sys.path.insert(0, "..")
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from ml.preprocessor import load_songs, CONTENT_FEATURES, build_feature_matrix
plt.style.use("dark_background")
df = load_songs("../data/processed/songs_features.csv")
print(f"{len(df)} songs loaded")
df[["title","artist"]+CONTENT_FEATURES[:4]].head()
```

## Feature Distributions

```python
fig,axes=plt.subplots(3,3,figsize=(14,10))
for ax,feat in zip(axes.flat,CONTENT_FEATURES):
    ax.hist(df[feat].dropna(),bins=30,color="#c9a84c",alpha=0.8)
    ax.set_title(feat,color="#c9a84c")
plt.tight_layout(); plt.show()
```

## Test Content-Based Model

```python
from ml.content_based import ContentBasedRecommender
model = ContentBasedRecommender().fit(df)
seed = df.iloc[0]
print(f"Seed: {seed["title"]} — {seed["artist"]}")
model.recommend(seed["id"],n=5)[["title","artist","content_score"]]
```

## Similarity Heatmap

```python
from sklearn.metrics.pairwise import cosine_similarity
X,_=build_feature_matrix(df)
S=cosine_similarity(X[:20])
labels=[r["title"][:12] for _,r in df.head(20).iterrows()]
plt.figure(figsize=(10,8))
sns.heatmap(S,xticklabels=labels,yticklabels=labels,cmap="YlOrBr")
plt.xticks(rotation=45,ha="right",fontsize=8); plt.tight_layout(); plt.show()
```
